```python
#FinDing Nearest Neighbours 
Distance=[]
name=[]
i=0
for testData in movies_data['genres']:
    dist=0;
    #print(len(testData))
    for x in range(0,len(sampleData)):
        if sampleData[x] not in testData:
                dist+=1
    #print(dist)
    name.append(movies_data.loc[i,'title'])
    i+=1
    Distance.append(dist)
df = pd.DataFrame(list(zip(name, Distance)),columns =['Name', 'Distances']) 
df
```
